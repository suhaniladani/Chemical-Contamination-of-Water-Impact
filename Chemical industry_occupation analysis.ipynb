```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
chemicals = pd.read_csv('chemicals.csv')
```

```python
IndusOccu = pd.read_csv('industry_occupation.csv', encoding="ISO-8859-1")
```

```python
earnings = pd.read_csv('earnings.csv', encoding="ISO-8859-1")
```

```python
earnings.head()
```

```python
earnings.describe()
```

```python
ChemEarning = pd.concat([chemicals, earnings], axis=1, join='outer', ignore_index=False)
```

```python
ChemEarning.head()
```

```python
ChemEarning.corr()
```

```python
test = earnings[['wholesale_trade', 'year']]
```

```python
WholeSale_Trade = test.groupby('year').size()
print(WholeSale_Trade)
#Plotting the Graph
plot_by_year = WholeSale_Trade.plot(title='Yearly Sales',xticks=(2010,2011,2012,2013,2014,2015,2016))
plot_by_year.set_xlabel('Years')
plot_by_year.set_ylabel('Wholesale Trade')
```

```python
test1 = IndusOccu[['total_employed', 'year']]
```

```python
import seaborn as sns
sns.set()
TotalEmpYear = test1.groupby('year').size()
print(TotalEmpYear)
#Plotting the Graph
plot_by_year = TotalEmpYear.plot(title='Yearly Employment',xticks=(2010,2011,2012,2013,2014,2015,2016))
plot_by_year.set_xlabel('Years')
plot_by_year.set_ylabel('TotalEmpYear')
```

```python
import seaborn as sns
sns.set()
chemicals.chemical_species.value_counts().plot(kind='bar')
```

```python
ChemEarningIndus = pd.concat([ChemEarning, IndusOccu], axis=1, join='outer', ignore_index=False)
```

```python
#correlation matrix
import seaborn as sns
corrmat = ChemEarningIndus.corr()
f, ax = plt.subplots(figsize=(19, 12))
sns.heatmap(corrmat, vmax=.8, square=True);
```
